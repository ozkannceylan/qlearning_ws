```python
df1=pd.read_csv('../Data/final_training_results.csv')
df1=df1.drop(columns=['alpha','gamma','t_init','t_grad','t_min','epsilon_init','epsilon_grad','epsilon_min'])
df1
                
```

```python
df2=pd.read_csv('../Data/final_test_results.csv')
df2
```

```python
merged_df = pd.merge(df1, df2, left_on=["type", "agent_init", "scenario"], right_on=["type", "agent_init", "scenario"])

merged_df
```

```python
merged_df.to_csv('../Data/final_results.csv', index=False)
df=merged_df
```

```python
# Sample data (replace with your actual values)

df1['scaled_steps'] = ((df['avg_steps'] - df['avg_steps'].min()) / (df['avg_steps'].max() - df['avg_steps'].min()))
df1['scaled_reward'] = (df['avg_reward'] - df['avg_reward'].min()) / (df['avg_reward'].max() - df['avg_reward'].min())
df1['scaled_last_10_reward'] = (df['avg_last_10_reward'] - df['avg_last_10_reward'].min()) / (df['avg_last_10_reward'].max() - df['avg_last_10_reward'].min())
df1['scaled_episode_time'] = (df['avg_episode_time'] - df['avg_episode_time'].min()) / (df['avg_episode_time'].max() - df['avg_episode_time'].min())


df1['scaled_success'] = (df['avg_success'] - df['avg_success'].min()) / (df['avg_success'].max() - df['avg_success'].min()) 
df1['scaled_obstacle'] = 1 - (df['avg_obstacle'] - df['avg_obstacle'].min()) / (df['avg_obstacle'].max() - df['avg_obstacle'].min())
df1['scaled_duration'] = (df['avg_experiment_duration'] - df['avg_experiment_duration'].min()) / (df['avg_experiment_duration'].max() - df['avg_experiment_duration'].min())

train = [df1['scaled_steps'], df1['scaled_reward'], df1['scaled_last_10_reward'], df1['scaled_episode_time']]
df['train_score'] = sum(train) / len(train)

test = [df1['scaled_success'], df1['scaled_obstacle'], df1['scaled_duration']]
df['test_score'] = sum(test) / len(test)

df['experiment_score'] = (df['train_score'] + df['test_score']) / 2

df
```

```python
scaled = ['scaled_steps', 'scaled_reward', 'scaled_last_10_reward', 'scaled_episode_time', 'scaled_success', 'scaled_obstacle', 'scaled_duration']
df = df.drop(columns=scaled)
df
```

```python
df.to_csv('../Data/final_results.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and is already loaded
def plot_exp_comparison(df, agent_init):
    scenarios=['Moderate',
    'FL-HE',
    'FL-LE',
    'SL-HE',
    'SL-LE',]

    epsilon_greedy_rewards = df[(df['type'] == 'epsilongreedy') & (df['agent_init'] == agent_init)]['experiment_score'].tolist()
    softmax_rewards = df[(df['type'] == 'softmax') & (df['agent_init'] == agent_init)]['experiment_score'].tolist()

    # Set width of bars
    bar_width = 0.35

    # Positions of the bars
    epsilon_positions = [x - bar_width / 2 for x in range(len(scenarios))]
    softmax_positions = [x + bar_width / 2 for x in range(len(scenarios))]

    # Create bars
    plt.bar(epsilon_positions, epsilon_greedy_rewards, width=bar_width, label='Epsilon-Greedy',)
    plt.bar(softmax_positions, softmax_rewards, width=bar_width, label='Softmax')

    # Labels, title, legend
    plt.xticks(range(len(scenarios)), scenarios, rotation=0)  # Set x-ticks in the center of each group
    plt.xlabel('Scenario')
    plt.ylabel('Experiment Score')
    plt.title(f'Experiment Score Comparison for {agent_init} initial position: Epsilon-Greedy vs. Softmax')
    plt.legend()

    # Grid for better readability 
    plt.grid(True, axis='y', linestyle='--')

    plt.tight_layout()  # Adjust layout to not cut off labels
    plt.show()

plot_exp_comparison(df, 'random')
```

```python
plot_exp_comparison(df,'constant')
```
