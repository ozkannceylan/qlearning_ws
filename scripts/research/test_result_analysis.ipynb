```python
path="../Data/softmax/Hyperparam_opt_logs_"
summary=[]
for i in range(1,16):
    result_path=path+str(i)+"/test_results.csv"
    data=pd.read_csv(result_path)
    test_results={}
    test_results["test_number"]=data["test_number"].max()
    test_results["success_rate"]=(data["success"]).mean()
    test_results["obstacle_collision"]=(data["obstacle"]).mean()
    test_results["experiment_duration"]=(data["experiment_duration"]).mean()

    summary.append(test_results)   
summary=pd.DataFrame(summary)
summary

# data
```
